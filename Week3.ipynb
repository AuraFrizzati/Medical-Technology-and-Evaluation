## **3.1: What is a Medical Technology Assessment?**

### **The Types and Uses of Assessments**

#### **Introduction to technology assessments**
- Valuation of technology informs important decisions during the entire life cycle of a medical technology
  - Go/No Go
  - Coverage Decisions
  - Pricing
- There are standard sets of tools to measure the value of medical technologies
- These sets of tools provide a systematic approach to valuing and ranking medical technologies

#### **Technology assessment basics**
- Technology assessment is the **collection**, **analysis**, and **dissemination** of information to relevant parties
  - **Government regulators** want to see evidence before approving the product
  - **Payers** want to see evidence of effectiveness or cost effectiveness before they will pay for it
  - **Physicians and hospitals** must be convinced to use/prescribe/buy the technology
- Need to collect the right information throughout the process
  - **Clinical data**
  - **Claims data**
  - Any **other relevant data**

#### **Common types of technology assessments**
- **Cost Analysis**: Only consider costs; least costly intervention is preferred
- **Cost-Effectiveness Analysis** (**CEA**): The costs of the intervention relative to some **outcome** in **natural units** (e.g. weight, mortality, or length of stay)
- **Cost-Benefit Analysis**: A **dollar value** is placed on the benefits as well as the **cost*** of the intervention 
- **Cost-Utility Analysis** (**CUA**): A **measure of quality of life** is incorporated with the **financial measures**

#### **Additional types of technology assessments**
- **Cost- Minimization Analysis**: Only costs are considered but can only be used when the **outcomes of the treatments are similar**
- **Burden of Disease Analysis**: Measures the impact of a disease on **mortality**, **morbidity** or **quality of life** across a population 
  - For example: malaria results in xx lost life years
  - There are multiple approaches to incorporating quality of life 

#### **General principles of any health technology assessment**
- Consider the **audience** for the assessment and the **perspective** of the analysis
  - Is it for the FDA or payer?
  - Is it from the patient or societal perspective?
- The assessment should be **unbiased** and **transparent**
  - Just because there is a specific audience does not mean there should be a specific result
- The assessment to should use as much **timely data** as possible
  - **All data sources** should be considered



### **Choosing the Appropriate Assessment Method**

#### **Which technology assessments is best**
- All use **standardized approach(es)** to measuring the net benefits of technology
- They are **comparison tools** to help evaluate choices
 - No one method will always indicate a clear choice, but it will provide a quantitative evaluation of the options

#### **Why choose cost-effectiveness analysis (CEA)**
- CEA is methodology to measure the **health benefits** attributable to a given intervention
- The health benefits are measure in **natural units** (e.g. LOS, amount/length of symptoms, time until death, etc)
- **Do not need information about financial implications of the outcomes**, which may not always be available
- Relatively **easy to explain** to many stakeholders
- The main limitation is that CEAs **do not provide a comprehensive assessment**, as the results are limited to the health benefits included in the analysis (usually only 1-2 health benefits are included)

#### **Why choose cost-benefit analysis (CBA)**
- CBA incorporates **dollar (or other monetary) values for the benefits** and can be helpful for making **investment or financial decisions** 
  - Compare the **total expected cost** of each option against the **total expected benefits** to determine whether the benefits outweigh the costs, and by how much. **Benefits** and **costs** in CBA are **measured in the same units**.
- CBA is a **more complete analysis of total benefits** than CEA but **requires more information/data/analysis** 
- CBA also requires applying dollar values to health benefits which **may not appeal to all audiences**

#### **Why choose cost-utility analysis (CUA)**
- CUA incorporates a **measure of health related quality of life** along with **Dollar values**
  - Common quality of life measure is a **quality adjusted life year** (**QALY**). QALY measures both **mortality** and **morbidity**.
- Comparisons of a **common measure** (**dollars/QALYs gained**)
  - Values health improvement for every remaining year of life after the treatment
- CUA methods are a **complete accounting of benefits** for patients compared to other methods
- **Societal benefits are not accounted** for and many people disagree with valuing life years

#### **Example: Uses of technology assessments**
- **National Institute for Health and Clinical Excellence** (**NICE**)
  - A UK organization that provides guidance on the use of new and existing medical technologies
- The National Health Service in England and Wales must cover treatments recommended by NICE
  - Is influential beyond NHS too
- NICE reviews many sources of evidence regarding new technologies and treatment techniques and performs technology assessments such as CEAs and CUAs
- NICE considers both the audience and the perspective of the decision to be made so to choose the most appropriate type of technology assessment.

### **A basic technology assessment**

#### **Identifying components for an assessment**
- Example: Gardasil - a human papillomavirus (HPV) vaccine 
  - Gardasil is a 3-dose HPV vaccine recommended for females age 9-26
  - Prevents cervical cancer
- One approach is to try to think of all the **costs** and **benefits** (to obtain a **cost/benefit ratio**)
- An alternative approach is to think of the **resources** and **effort** required **with or without treatment**
  - **Health care resources**
  - **Non-health care resources**
  - **Time**

#### **Identify health care resource use** 
- There are two sources of costs:
  - The vaccine cost
  - 3 office visits to deliver the vaccine are health care costs that should be included
- Because it prevents cancer, cost savings should be considered too

#### **Identify non-health care resources**
- Costs of transportation to and from the hospital or doctor’s office to get the shots
- Savings from avoided costs for travel for cancer treatments

#### **Costs to patients**
- **Patient time spent in treatment** included in numerator: a 26-year-old woman who leaves work to obtain the 3-doses of vaccine, the resources used would include:
  - Time traveling to and from the doctor
  - Time waiting to receive vaccination
  - Time spent in the actual office visit receiving the vaccine
  - This time also has an opportunity cost

#### **Identify other sources of costs**
- Informal caregiver time, stress (e.g. related to deal with someone who has received the vaccine), etc.
- In the case of Gardasil the costs from the time and effort of informal caregivers could include:
  - A parent might have to leave their job early to take their daughter to the doctor to get the vaccines, 3 times, once for each dose
  - If the daughter has an adverse reaction, the parent might have to stay home to care for her
  - The costs may be just as high or higher if a daughter is diagnosed with cancer


## **3.2: Approaches to measuring costs**

### **Introduction to costs**
#### **Getting started…**
- Why do we care about costs in health care? 
  - **Efficiency**: **limited resources** means that the more resources used producing health care, the less we have for other things
- What do we mean by the **cost** in a **CEA/CBA/CUA**? 
  - **Prices** are the most obvious example (if there is a competitive market)
  - Generally “cost” is the **marginal cost** or **“opportunity cost”** of 
the resources (the next best use of a resource). E.g. time-cost.

#### **Types of costs**
- Costs are generally divided into two categories: **Direct** and **Indirect**
- **Direct costs** are those that be **attributed to the intervention**
  - Drugs, tests, supplies, healthcare personnel, and medical facilities, etc.
- **Indirect costs** are those that **result from the use of the intervention**:
  - Travel time, emotional effects, caregiver time and effort, etc.

#### **Cost perspectives**
- **Patients**: out of pocket costs, inconvenience of treatment.
- **Hospitals, Doctors, etc.**: costs of providing care or the payment/reimbursement for providing care
- **Insurers**: payments they make to providers and the costs of processing claims 
- **Government**: Costs of paying for care, costs to other government services, national/state/local productivity measures
- **Society**: concerned with all resource

#### **Why Measure Costs?**
- Could compare only benefits OR relate benefits achieved to the resources needed to understand the value of the benefits 
- Example: new drug to prevent heart disease is approved
  - What will insurers want to know before they cover it?
    - Potential benefit
    - Expected costs

#### **Three Steps to Measuring Costs**
- **Identify** types of **resources** to be included (the **analysis perspective**)
- **Measure** the **resources** 
- Place a **monetary value** on all **resource measurements**, sum those values and identify total costs


### **Micro and gross costing methods**
#### **Approaches for determining costs**
- **Micro-costing methods** disaggregate health care utilization into **fundamental units**
  - Example Myocardial infarction treatment includes: 
    - the ER, 
    - the physicians and techs in the ER, 
    - hospital room, 
    - nursing care, 
    - food, 
    - follow-up physician visits, 
    - cardiac tests, 
    - drugs, etc.
- **Gross-costing methods** aggregate health care utilization into **larger collections of resources** (often on an **expected basis**)
  - Example: Hypertension costs include:
    - physician visits, 
    - a nursing home stay, 
    - DRGs payment for stroke, myocardial infarction, coronary arterial bypass graft procedure, angina, etc

#### **Micro-costing methods**
- Requires **closely examining** the **actual resources** used by a **particular patient**
  - **Electronic medical records**, **chart reviews**, **itemized bills** are potential sources of micro-level cost data
  - There is **large variance** in the ability of providers capture micro-level data

#### **Gross-costing methods**
- Can use the **reported or standardized payment** as the cost
- Many sources of **aggregated data**
  - **Administrative claims data** provide estimates of gross utilization
  - **Direct patient surveys** are also a source of gross-cost data
    - Child care time
    - Informal caregiver time
    - Travel time
  - **Literature** can provide gross costing estimates as well 
- Often times **more feasible** than micro-costing but may be **less accurate than micro-costing**

#### **Calculate total costs**
- Once utilization is measured and costs are identified, **unit costs** need to be used to determine total costs
- **Prices** are the most common means of measuring unit costs
  - Assumes that price accurately represents represent the marginal cost, or the opportunity cost of the resources used




### **Unit costs**

#### **Considerations unit costs** 
- Hospital **charges** (e.g. charge listed for an intervention, \$200) are not the same as **prices** (how much the insurance pays for the intervention, \$150) or **costs** (actual cost of the intervention, \$125)
  - **Hospital charges** can be **adjusted to reflect costs** by using the Medicare cost to charge ratio found in Medicare Cost Reports
  - Hospital prices for privately insured patients need to be calculated using administrative claims data from a commercially insured population
- **Price inflation** can be **adjusted** for by using the **Consumer Price Index** (**CPI**) or the **Medical Component of the CPI** (**MCPI**)
  - The MCPI is based specifically on medical care products and services
- **Patient time cost** should be estimated by the **wage rate** of the patients targeted by the intervention
  - **Stay at home parents wages** need to be imputed
    - Find the age, gender, education, labor experience of the housewife and impute the wage from a regression equation that estimates wage from these variables for the entire population
  - **Wages for retired people** and **children** receiving health care could also imputed wage
    - Because they are not working there is no productivity time costs, but there are leisure time costs and the wage rate reflects the value of leisure, too 
- **Informal caregiver time** also needs to be imputed
  - Again, find the age, gender, education, labor experience of the 
informal caregiver and impute the wage from a regression equation that estimates wage from these variables for the entire population
- Example: what about person who stays home from her job as an attorney to care for an ailing parent?
  - Could use her actual wage as a proxy for the unit cost of caregiving

#### **What about future costs**
- Generally **future medical/health care costs** should be included
   - These costs are **not controversial** if they are **associated with the condition being treated by the intervention**
- Future medical/health care **costs unrelated to the intervention** or condition are **controversial**
  - These refer to medical costs due to treating another condition that 
resulted from treating the original condition (e.g. future costs for treating heart attack when evaluating a technoogy treating cancer)

## **3.3: Approaches to measuring effectiveness**

### **Benefits**
#### **Types of benefits**
- **Medical benefits** (they can all be measured directly)
  - Fewer side effects 
  - Faster recovery
  - Reduced mortality/longer life
- **Financial** (they can only be measured indirectly)
  - Avoided medical costs
  - Increased productivity
- **Combination**
  - Monetized value of extra life
  - Quality adjusted value of extra life

#### **Value of life measures**
- Uses **two commonly accepted outcome measures** to assess **benefits**:
  - **Years of life**
  - **Money**
- It can be a **contentious issue** to put monetary value on a life or life year
  - There are also numerous methods to assign value to life and widely ranging estimates
- These approaches **do not consider non-market factors which may reduce value** (e.g. pain and suffering)
  - Quality adjusted life years do though…

#### **Human capital approach**
- Equates the **value of life** to the **market value of the productivity** (output) **of a person's expected lifetime**
  - Discount the value of future earnings resulting from extension of life to their present value
- One main concern is **labor market imperfections** 
  - Different groups of people may be paid differently (e.g. **racial or gender bias**) 
- **Does not consider non-market productivity** (e.g. leisure)

#### **Willingness to pay approach**
- Based on **amount people are willing to pay to reduce probability of death**
  - Not using seatbelts, no smoke detector, smoking
- Assumes that the value of a person’s life is equal to the **probability 
of a reduction in death divided by the cost of the intervention**
  - Gives a lower bound of the value of life 
  - Value is usually higher than human capital approach because it includes 
market and non-market factors



### **Quality Adjusted Life Years (QALYs)**

#### **Quality adjusted life years (QALYs)**
- Quality-adjusted life years (QALYs) **combine morbidity and mortality** into a single measure
  - **Mortality** is measured in life years or **expected life years**
  - **Morbidity** measured by **quality of life** (**QoL**)
- **Quality-of-life dimensions** are **weighted** according to **individual preferences** measured in units known as **“utility”** 
  - Quality-of-life weights range from 0 to 1 where 0 is the quality of 
life associated with death
- The **QALY = (QoL measure) x (Number of years a health state)**

#### **QALY caveats**
- First, it is important that **death** has **QoL weight of 0** in order to be able to combine morbidity and mortality
  - There are some who argue that **living with some disease is worse than death** this would be represented by a **negative QoL**
- Second, the **quality-of-life scale** must have **interval property**
  - An increase from 0.3 to 0.4 has the same meaning in terms of quality of life improvement as from 0.7 to 0.8

### **Measuring Quality of Life (QoL)**
#### **Quality of Life (QoL) measurement**
- In order to calculate QALYs we need to **identify patient’s utility** in **health states**
- A **health state** is the health of an individual at a **particular point in time** 
  - **Health-related quality of life** (**HRQoL**) **measures** are used to define health states
- **Large surveys** of the views, attitudes, and preferences of the patients and the general public are used to assign utilities to health states

image.png


#### **HRQoL survey instruments**
- **General QoL survey**: **Short Form Health Surveys** are the most common (**SF-36** and **SF-12**)
  - Computer programs exist to convert scores to SF-6D and utilities
- Many **disease specific QoL surveys**
  - Very specific to intended effects of treatments
  - Can be sensitive to small changes in disease activity/status
  - Clinicians can link to specific actions
- Can also collect QoL measures during clinical trials

#### **QoL information collection in clinical trials**
- Could **include a measurement of QoL survey** to find out how many QALYs each treatment produces
- Have each patients **fill out the survey prior to treatment and after** they receive the treatment (or at various intervals pre and post follow-up)
  - Convert the HRQoL measures to QALYs
- With QALYs for each individual in the trial you can measure QALYs gained on average as a result of the treatment
