## **4.1: Conducting a Cost Effectiveness Analysis (CEA)**

### **Uses of Cost Effectiveness Analysis (CEA)**

#### **Example: Percutaneous coronary intervention**
- Assess whether percutaneous coronary intervention (PCI) 
with intensive pharmacologic therapy and lifestyle 
intervention (optimal medical therapy) is superior to optimal 
medical therapy alone in reducing the risk of cardiovascular 
events in patients stable coronary artery disease
- The trial found no significant differences between the PCI 
group and the medical-therapy group in the aggregate 
outcome measure of death, myocardial infarction (MI), and 
stroke.

#### **Example: A simplified medical technology assessment** 
- Standardized approach(es) to measuring the net benefits of 
technology
  - Measure the health costs and benefits attributable to a given intervention
- A basic understanding of the context of the study is needed 
before beginning an assessment: 
  - What is the technology being assessed? PCI
  - What is the patient population? Patients with stable 
coronary artery disease
  - What costs are involved with the competing technologies? 
The difference in cost is the cost of the PCI (it is the only difference between the 2 interventions)
  - What are the outcomes in the study? The health outcomes 
were death, MI, and stroke
  - Who is the audience? This type of study may be of interest 
to providers and insurers

#### **Example: Follow-ups**
- Additional considerations for the assessment
  - What regulatory agencies are involved?
  - What reimbursement policies are involved?
- Additional considerations following the assessment
  - How might insurers interpret these results? 
  - How might providers (doctors and hospitals) interpret these results? 

#### **Why not always use CEA?**
- CEA may be used in a situation where it is **clear that 
achieving a certain effect is desirable** because of its known 
impact on a specific health outcome
  - For example, a physician who treats diabetes patients might simply be interested in lowering hemoglobin A1c as cost-effectively as possible
- **CEA cannot be used to allocate spending in a budget 
because various diseases have diverse measures of effectiveness and are not comparable** (in these situations, more complex analyses such as CBA or CUA might be used)
  - Hemoglobin A1c, cholesterol, and pain measures are not directly comparable








### **Constructing an Incremental Cost Effectiveness Ratio (ICER)**

#### **The Incremental Effectiveness Ratio (ICER)**
image.png

- ICERs are the **costs needed** to obtain **one unit of benefit** (or **effectiveness**) (measured in costs per benefit unit, e.g £/QALYs)
- Measures the benefits associated with an intervention relative to its cost

#### **Reference Intervention**
- Calculation of an ICER requires a **reference intervention** or **comparator technology**
  - Usually, **current medical practice** or the **standard of care**
  - For example ACE inhibitors vs. Diuretics to treat high blood pressure
- It is necessary to distinguish between **independent** and **mutually exclusive interventions**
  - Medical therapy and lifestyle changes are independent but can be used concurrently
  - A bare metal stent and a drug eluting stent can not be implanted at the same time to treat the same condition (mutually exclusive interventions)

#### **Numerator of the ICER**
- The **cost of the intervention** is **all the costs associated with providing the treatment** 
  - Hospitalization, time, drugs
- **Costs that are included in the denominator do not get included in the numerator**
  - The costs avoided by an intervention include avoided future health care costs

#### **Denominator of the ICER**
- Denominator includes a measure of the **benefits of an intervention**
  - **Health related** (e.g. symptom free days, mortality)
  - **Financial** (e.g. costs avoided)
  - **Quality of life** (e.g. quality adjusted life years)
- **QALYs are commonly used** because the take into account **longevity** and **quality** thus are considered a more 
comprehensive measure

#### **What makes an intervention cost effective?**
- Need to assign a value to a life or year of life?
- There is no single cutoff for cost effectiveness
  - Roughly, in the U.S. anything between \$50,000 to \$100,000 per QALY is a “good buy”

### **Interpreting an Incremental Cost-Effectiveness Ratio (ICER)**
- The ICER can also be **expressed graphically** on the  **“Cost-Effectiveness Plane”**

- **Cost** is measured on the **y-axis**
- **Effectiveness** is measured on the **x-axis**
- The **ICER** is equal to the **slope** of the line on the cost-effectiveness plane
- The **origin** of the plane is the reference intervention

## **4.2: Advanced methodologies**

### **Bayes' rule**
#### **Incorporating new information**
- The information you want is not always the information you 
have
- How should one **rationally incorporate new information into their beliefs**?
  - For example, suppose one gets a positive test result (where the test is imperfect), what is the probability that one has the condition?
- Answer: **Bayes’ Rule**!
 - Particularly useful for the analysis of screening but it applies more broadly to the incorporation of new information

#### **Bayes’ Rule**
- Notation:
  - **P(A** = Probability of event A (unconditional)
  - **P(B)** = Probability of information B occurring
  - **P(B|A)** = Probability of B occurring if A (conditional)
  - **P(A|B)** = Probability of A occurring given information B (This is the object we are interested in)

- What is the probability that a woman has breast cancer?
  - 1% of women at age forty who participate in routine screening have breast cancer (P(A)=0.01)
  - 80% of women with breast cancer will get positive mammograms (P(B|A)=0.8)
  - 9.6% of women without breast cancer will also get positive 
mammograms (P(B|negA)=0.096)
  - A woman in this age group had a positive result in a routine screening

### **Discounting (or accounting of time)**
- The **impact** of most medical technologies **stretches over time**.
- The **future** is **“worth less”** than the present.
- We can calculate the **present value (NPV) of future costs** in (**time T**) using a **discount rate** (**r**):

image.png

#### **How much does it matter?**
- It depends… 
  - May matter for **disease screenings** and many **preventions** 
- **Time implications** are important if there are **long time frames ** and/or if the **intervention is marginally cost effective**.
- Suppose an intervention increases health related quality of life by 0.2 units for 10 years.
  - Assume QALYs = \$100,000.
  - Depending on the discount rate the discounted present value of a QALY can differ a lot:
0%: \$200,000, 3%: \$175,722, 5%: \$162,156, 10%: \$135,180

#### **What discount rate?**
- There are **many possible discount rates** – there is not one correct one
  - The discount rate used should **capture the rate at which society is willing to trade off current for future consumption**.
- **Market based discount rates** (**based on interest rates**) are common
  - Advantage of using market rates is that there is a theoretical justification 
  - Disadvantage of using market rates is that they may not reflect the type of transactions that are relevant for health care.
- **Rates of 3% to 5% are reasonable** and consistent with other 
studies 
  - May want to test implications of rates between 0% and 7%

#### **Discounting QALYs**
- There is some **controversy** about how (and if) to discount QALYs
- Is a future life year worth less than a contemporaneous one?
  - The prevailing practice is to use the same discount rate for costs as one uses for health consequences 
  - If cost and health benefits discount rates are different there are implications for implementing the intervention 
- If future health worth more you want to delay interventions 
- If is worth less you want to ‘over implement’ today

#### **Other issues with discounting**
- Should we use a constant discount rate?
  - It is **usually better to use a constant discount rate** than a variable discount rate 
- Keep valuations and discounting separate!
  - **Apply discounting after costs and benefits are calculated** in order to identify the impact of discounting
- What about **inflation**?
  - If combining expenditures from multiple years the analysis should be conducted in ‘real’ terms

### **Decision models**

  - There is a risk of infection from needle-stick injury
  - Prophylaxis treatment can be given to prevent HIV infection, but there are many potential side effects

#### **Evaluating the model**
- Apply **probabilities** to **each possibility** in the decision model and **outcome measures** to each **end point**
  - At **each split** in the decision tree **the values have to add up to 1 (proportions) or 100% (percentages)**
- **Probabilities** and **outcome measures** should come from **good quality research evidence**
- The **probabilities times the outcome measure** equals the **expected value of the outcome**

#### **Identifying the Probabilities**
- Average risk for HIV transmission after exposure is approximately 0.3%
- Effectiveness of prophylaxis is difficult to estimate 
  - A case control study indicated that prophylaxis reduced odds of HIV infection by 81% which is equivalent to a 5% chance of HIV if prophylaxis is taken
- Side effects include nausea/vomiting, malaise/fatigue, headache, myalgia, abdominal pain, diarrhoea 
  - Probability of getting a side effect 50-75% (assume 63%)

- Existing preference scores associated with HIV infection and some side effects based on exiting literature
  - Preference scores for HIV range from 0.5 – 0.75
  - Nausea and vomiting – 0.9863
  - Diarrhoea – 0.81
  - Abdominal pain – 0.9863
- Values used in model 
  - HIV infection 0.5
  - side effects 0.9
  - no infection 1.0
  - Infection and side effects 0.4

#### **Results**
- The expected value of no prophylaxis is .99715
- The expected value of prophylaxis is .9369
- Based on this model and the assumptions made a person is better off with the no prophylaxis treatment option

## **4.3: Incorporating uncertainty**

### **Introduction to uncertainty**
#### **Uncertainty in technology assessment models**
- There are **several sources of uncertainty** that may impact results
  - **Statistical uncertainty** (we might know the average impact but some specific cases will be higher or smaller than the average)
  - **Randomness** in **costs** and **outcomes**
- Analyzing the **how results** of technology assessments vary  **for alternative parameter values** is called **sensitivity analysis**

#### **Statistical Uncertainty**
- The **main source of uncertainty** in **assessments** and **models** is from **parameter uncertainty**
  - Not every occurrence of an event is exactly the same:
  - An intervention may reduces chance of heart attack by 12% with 5% error
  - The cost of intervention is \$20,000 with a standard deviation of \$4,000
- As with all other aspects of the assessment it is important 
to use **reliable information about the distribution** of **outcomes** and **costs** for the sensitivity analyses

#### **A sensitivity analysis example**
- How much would a decrease in Medicare reimbursement for bypass surgery impact a hospital’s profit margin?
  - Reimbursement is \$120,000
  - Reimbursements are expected to decrease 2.5%
  - Average costs are \$110,000 (the standard deviation is \$45,000)
- May want to calculate the impact using 2%, 3%, and 3.5%
- May also want to calculate with plus/minus one standard 
deviation from the average

#### **Types of sensitivity analyses**
- **One-way sensitivity analysis** changes **one variable at a time**
  - First estimate the impact for each possible change in reimbursement then for each possible cost
- **Multi-way sensitivity analysis** changes **more than one variable at a time**
  - Calculate the impact of profit margins for different levels of costs and changes in reimbursement concurrently

### **Monte Carlo Simulation**

#### **Monte Carlo Analysis**
- **Monte Carlo** (**MC**) **simulation methodology** used to examine the impact of uncertainty
- The **parameters of interest** in an analysis will have a **distribution**
  - The **average benefit or mean** (30 symptom free days) may be from normal **distribution of benefits** which also has a variance 16 days
- MC methodology requires **random draws from a sample based on the parameter distribution**
  - Use a **random number generator**

#### **Performing the Monte Carlo analysis**
- **Randomly draw** a **parameter value** from the **known distribution**
  - Draw a value of symptom free days from a Normal distribution with a mean of 30 and a variance of 16 days
- **Perform the analysis using the randomly selected parameter** value and save the output
  - Calculate an ICER using the randomly selected parameter value of 22 days
- **Repeat the previous two steps numerous times** (one hundred to one million)
- **Combining the output of each calculation will result in a 
distribution** 
  - The **distribution of the ICERs** can be used to determine the **mean and variance of the CE (cost-effect) result**

#### **A Monte Carlo analysis example**
- A new cancer screening blood test for a certain types of cancer has been developed
- If levels of a protein found in the blood are higher than levels in the 99th percentile of a healthy population more expensive testing is necessary 
  - The 99th percentile protein level of the healthy population is 0.7
- There are two brands of tests available:
  - One significantly cheaper than the other but has more variation in the result (They have the same average but different variances - the coefficient of variation is .25 compared to .1) 
- If a test result is positive the additional course of testing costs \$1500 

#### **Performing the Monte Carlo analysis**
- Assume the true protein level value is 0.7
- Randomly draw a test parameter result from the known distribution for each test
- Calculate the expected value of treatment using the the
randomly drawn parameters
- Repeat the previous two steps 1,000 times
- From the 1,000 results for each test calculate the average cost implications of each test

### **Markov Models**

#### **Markov Modeling**
- Markov models are a methodology for **modeling uncertain future events**
  - They are **similar to decision models** but Markov models **incorporate uncertainty into decision points** in the model 
  - Allows for **changes in the progression of diseases overtime** by assigning subjects to differ health states
- In a Markov model the **probability of being in one state depends on the previous state**
- The models are **often calculated using Monte Carlo methods**

#### **A Markov modeling example**
- Assume there are three possible treatments for cancer:
  - Chemotherapy
  - Surgery 
  - Chemotherapy and surgery
- Following each treatment there are four possible outcomes:
  - No cancer occurrence
  - Local cancer occurrence
  - Metastases (i.e. cancer spreads)
  - Death
