## **2.1: Clinical Trial Design**

### **Intro to clinical trial**
We discuss what clinical trials are and how they can be useful in collecting data for a technology assessment.

#### **Developing a Research Question**  
Example – assume want to know more about the cost 
effectiveness of a once per year rheumatoid arthritis (RA) 
treatment?   
--> **Defining the research questions** helps focus and properly define what it is we really want to know

**PICO protocol**
- P: Population
- I: Intervention
- C: Comparator(s) (e.g. once per year treatment vs once per month treatment)
- O: Outcomes(s) (e.g. measures of side effectc, improvements, mobility, function, other outcomes specific for rheumatoid arthritis that are reliably measureable)

#### **Types of Clinical Studies**
- **Non-Experimental** (**Observational**)
  - **Observe** both **exposures and outcomes** (E.g. **Case studies**, **Prospective observational studies**); e.g. one population receiving the intervention and another one not receiving it.
  - Identify populations, observe and record characteristics, look for associations
  - A good way to describe a non-experimental trial is the
researcher CANNOT control, manipulate, or alter the intervention, compactor, or population but instead relies on interpretations, observations, or interactions to come to a conclusion.

- **Experimental** 
  - **Assign exposure** & **observe outcomes**: (E.g. Randomized controlled trials (**RCTs**))
  - Identify participants
  - Place in **common context & intervene**, observe/evaluate effects of intervention

#### **Clinical Trials**
- Clinical trials provide the ability to **reduce bias** and **variability** in order to identify the effects of treatment or intervention
- **Bias** affects **accuracy**. Accuracy is how close to a measured value is the true value
- **Variability** affects **precision**. Precision is how close to a measured values are to each other

### **The role of randomisation**
#### **Ensuring comparability of treatment groups**
- Randomisation helps to balance all factors between treatment groups
- If randomisation "works", then groups will be similar in all aspects except for the treatment received
- Allocation of treatments to participants is carried out using a **chance mechanism** so that neither the patient nor the physician knows in advance which therapy will be assigned
- Randomisation prevent biases occurring in the treatment, although variation can still be present

#### **Alternatives to Randomization**
- Randomization is not always possible due to ethical or 
practical considerations
- Some alternatives:
  - Historical controls (how people where responding to treatment before introduction of new technology)
  - Non-randomized concurrent controls
  - Different treatment per physician
  - Systematic alternation of treatments
- Sources of bias for these alternatives need to be 
considered

#### **RCT**
- Advantages: 
  - Unbiased distribution of confounders 
  - Blinding more likely 
  - Randomization facilitates statistical analysis (e.g. determining statistical accuracy of the intervention effect)
- Disadvantages: 
  - Expensive: time and money
  - Volunteer bias 
  - Ethically problematic at times

#### **Example: Randomizing patients to RA treatments**
- Identify patient population and clinical sites
- Doctors or administrators talk with eligible patients to determine if they are eligible
- Once a full sample of patients is identified use a random number generator to assign patients to groups
- Create a patient list in Excel and use the random number generator function to create random number between 0 and 1 for each name
- All patients with a number less than or equal to .5 are in one group. 
- The others are in the other group
- Can’t rely on human selection or patient selection into random groups

### **Blinded designs**
They allow to reduce bias so the outcomes of the study can be used to technological assessment.

#### **Blinding is an important element of trial design**
- Blinding is masking the identity of the assigned 
interventions
- The main goal is to avoid potential bias caused by 
conscious or subconscious factors
  - **Single blind**: patient is blinded
  - **Double blind**: patient and assessing investigator are blinded
  - **Triple blind**: committee monitoring response variables (e.g. statistician) is also blinded

#### **Why should patients be blinded?**
- Patients who know they are receiving a new or 
experimental intervention may report more (or less) side 
effects
- Patients not on new or experimental treatment may be 
more (or less) likely to drop out of the study
- Patient may have preconceived notions about the benefits 
of therapy
- Patients try to get well/please physicians

#### **Why should investigators be blinded?**
- The treating physicians and investigators are often the 
same people - Possibility of unconscious bias in 
assessing outcome is difficult to rule out
- Decisions about concomitant/compensatory treatment 
are often made by someone who knows the treatment 
assignment and “Compensatory” treatment may be 
given more often to patients on the protocol arm 
perceived to be less effective – different treatment 
protocols are then being compared

#### **How to Blind**
- A placebo treatment or intervention can be used to blind 
the patient
  - May also be work to blind the doctor
- Examples:
  - Pill of same size, color, shape as treatment
  - Sham operation (anesthesia and incision)
- Note: there are ethical considerations to placebo 
treatments

#### **When blinding is not possible**
- Standardize treatment groups
- Use objective reliable outcomes
- Consider duplicate assessment
- Acknowledge limitations

### **2.2: Administrative (claims) data**

#### **What is administrative data**
- Administrative data or health insurance claims are 
data from the billable interactions (i.e. financial data, not primarily clinical!) between: 
  - patients and the health care provider and 
  - health care provider and the payer 
- Standardized data that can be used to assess 
spending and utilization of health care services

#### **Intro to Claims Data: Medicare**
- Part A – hospital care, skilled nursing facility care, nursing home care, hospice, and home health services
- Part B – physician visits, ambulance services, durable 
medical equipment, mental health, preventative services
- Part D – prescription drug coverage

#### **Intro to Claims Data: Commercial insurance**
- Medical 
  - Inpatient: Hospital care, skilled nursing facility care, nursing home care, hospice, and home health services
  - Outpatient: Ambulatory clinics, outpatient surgery centers, facility fees
  - Professional: Doctor and allied health professional fees 
- Pharmacy
  - prescription drug coverage

#### **Claims vs. medical records**
- **Claims data**
  - Information from all doctors/providers caring for a 
patient
  - Population is insured only
  - Can identify diagnosis and procedures
  - Limitations: 
    - limited population, limited follow up (patients can move to another insurance), only billed 
services (e.g. over the counter medications are not included), lags

- **Medical records**
  - Only the portion of care provided by doctors/providers using the EMR (Electronic Medical Records)
  - Uninsured and insured patients
  - Can identify lab results, vital signs, free text notes
  - Lack of standardization, inconsistent data entry, single site of patient care 
  - Data from same patient needs to be linked from multiple EMRs




### **Elements of claims data**

#### **Claims Data: What is included**
- Membership information
- Dates and place of service
- Diagnosis 
- Procedure or drug codes
- Minimum provider information

#### **Claims Data: Codes**
- **International Classification of Diseases, V9** (**ICD-9**)
  - 5 digit, first digit alphanumeric, others numeric
  - Transitioning to **ICD-10** (seven digits, all alphanumeric 68,000 codes)
- **Procedure** - **Healthcare Common Procedure Coding System** (**HCPCS**)
  - Level I **Current Procedural Terminology** (**CPT**) - numeric
  - Level II – services not covered by CPT codes – alphanumeric
- **NDC** - Food and Drug Administration’s Nation Drug 
Code directory **bold text** (Drugs) 
  - 10 digit number with 3 segments: Labeler, Product, Package,

#### **Diagnosis related groups (DRGs)**
- Used for Medicare inpatient prospective payment (Part A)
- ICD-9 and CPT codes associated with the hospitalization 
episode are processed through “grouping” algorithms to 
result in a single DRG for payment from CMS.
- The position of codes matters for payment.

#### **Intro to claims: Financial elements** 
- **Costs** – amount it costs to produce/provide a service. They are not included in claims data though.
- **Charges** – amount billed by provider (i.e. the list price)
- **Payment** (allowed amount) – amount paid by the payer and insured out of pocket through
  - Deductibles,
  - Coinsurances, and
  - Co-pays
- **Expenditures** – **Payment** x **Quantity** = **Total spending**

### **Sources of claims data**

#### **Source of Claims Data**
- Commercial Claims (Optum, Truven - MarketScan)
- Medicare 
  - Limited data sets (LDS) 
  - Research identifiable files (RIF)
  - USRDS (ESRD)
- Medicaid (Medicaid Analytic Extract – MAX)
- Healthcare cost and utilization project (H-CUP) 

#### **Access to Data**
- Medicare & Medicaid:
  - RIF files only for research institutions (Research Data Assistance Center (ResDAC))
  - Aggregate-level data through private research groups that use CMS with approval 
  - LDS available to commercial entities to evaluate policy
- Commercially-insured claims data:
  - Data updated monthly to annually
  - Does not include full commercial claims population
- State all payer claims databases (APCDs)

## **2.3: Using data**
### **Bias in data**

#### **What is a bias in a clinical trial?**
- Bias: any influence which acts to make the observed 
results non-representative of the true effect of therapy 
- Examples:
  - Healthier patients given treatment A and sicker patients given treatment B
  - Treatment A is “new and exciting” so both the physician and the patient expect better results on A

#### **Types of bias in a clinical trial**
- **Reporting Bias**: Systematic differences between reported and unreported findings
- **Selection bias**: Systematic differences between baseline characteristics of the groups that are compared
- **Performance bias**: Unequal provision of care apart from treatment under evaluation
- **Detection bias**: Systematic differences between groups in how outcomes are determined. 
- **Attrition bias**: Biased occurrence and handling of deviations from protocol and loss to follow-up

#### **Variability in a clinical trial**
- Variability is different than bias
  - Variability is not systematic
- High variability makes it more difficult to discern differences in treatment effects
- What are some examples/sources of variability?
  - Measurement
  - Biologic
- It is not always possible to control for all sources of 
variability

#### **Reducing bias and variability in clinical trials**
- Good clinical trial design!
- Groups must be alike in all important aspects and only 
differ in the treatment each group receives

#### **What is bias in administrative data?**
- Bias can result from using **inappropriate data**
  - Example: using Medicare data to assess prices in the commercially insured population
  - Example: commercial claims from a unique employer/populations – using tech start-up claims from central California not the same as rural north east population
- Bias can also result from **improper controls** 
  - Prices and risk differ geographically and over time
- Note – claims data are themselves variable because there 
is some randomness to who gets sick and how severely

#### **Reducing bias in studies using claims data**
- Good **research design**!
- Understand the **strengths** and **limitations** of the data and what population is represented
- Apply **appropriate adjustments** for prices and severity

#### **Note about bias**
- No study is perfect and free from bias
- It is necessary to systematically check that the 
researchers have done all they can to minimize bias
  - Know the sources of any remaining bias and understand the impact of the bias


### **Combining trial data**
#### **Systematic Reviews**
- Systematic Review: is a literature review that tries to 
identify, appraise, select and synthesize all high quality 
research evidence relevant to a specific research question
- Independent study selection, data extraction, critical appraisal
- A priori selection criteria
- Transparent and reproducible

#### **Components of a systematic review**
- Formulation of a review question and inclusion criteria 
- Preparation of a protocol
- Identifying research evidence 
- Study selection 
- Data extraction 
- Quality assessment 
- Data synthesis (possibly using meta‐analysis)
- Presentation of results
- Interpretation and drawing conclusions

#### **Meta-Analyses**
- A meta-analysis is a statistical method used to aggregate 
the results of separate, independent studies in order to 
obtain an overall estimate of effect
  - Systematic reviews may contain meta-analysis
- Meta-analyses requires a variety of statistical methods:
  - Effect size measures transformations
  - Direction and magnitude of effect Heterogeneity
  - Publication bias 
  - Quality assessment and sensitivity analyses 
  - Considers variance within studies

#### **Question a Meta Analysis Can Answer**
- What is the **direction** of the effect? 
- What is the **size** of the effect? 
- Is the **effect consistent** across studies? (**heterogeneity**) 
- What is the **strength of evidence** for the effect? (**quality assessment**)
